```python
Amount = [data.loc[:,"Difference Between Expected Charges and Billed Charges (Rs.)"].where(data.loc[:,"Difference Between Expected Charges and Billed Charges (Rs.)"]==0).sum(axis=0),data.loc[:,"Difference Between Expected Charges and Billed Charges (Rs.)"].where(data.loc[:,"Difference Between Expected Charges and Billed Charges (Rs.)"]<0).sum(axis=0),data.loc[:,"Difference Between Expected Charges and Billed Charges (Rs.)"].where(data.loc[:,"Difference Between Expected Charges and Billed Charges (Rs.)"]>0).sum(axis=0)]
Amount
```

```python
st = ["Total orders where X has been correctly charged","Total Orders where X has been overcharged","Total Orders where X has been undercharged",]
st
```

```python
df = pd.DataFrame({'':st,'Count':count, 'Amount(Rs.)':Amount},columns=['','Count', 'Amount(Rs.)'])
df
```
